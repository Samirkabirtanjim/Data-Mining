**Distribution Plot**

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("/kaggle/input/depression/Depression Student Dataset.csv")
dataset
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=dataset, x='Gender',hue="Depression",palette='pastel')
plt.title("Gender vs Depression")
plt.show
```

**Box Plot**


```python
plt.figure(figsize=(8,8))
sns.boxplot(data=dataset, x='Gender', y='Age',palette='coolwarm')
plt.title("Box Plot of Age by Gender")
plt.show()
```

**Violin Plot**

```python
plt.figure(figsize=(8,8))
sns.violinplot(data=dataset, x='Gender',y='Study Hours',palette='muted')
plt.show()
```

**Line Plot**

```python
plt.figure(figsize=(8,8))
sns.lineplot(data=dataset, x="Gender",y="Age")
plt.show()
```

**Scatter Plot**

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=dataset,x='Gender', y='Age', hue='Depression')
plt.show()
```

**Pair Plot**

```python
sns.pairplot(data=dataset, hue="Study Hours", palette="husl")
plt.show()
```

Multivariate Bar Plot

```python
sns.catplot(data=dataset, x="Gender", y="Age", hue="Depression", kind="bar", ci=None, palette="pastel")
plt.title("Multivariate Bar Plot")
plt.show()
```
