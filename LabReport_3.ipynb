```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(52)
NumOfTransactions = 100
items = ['A', 'B', 'C', 'D', 'E', 'F']
Transactions_data = np.random.randint(2, size=(NumOfTransactions, len(items)))
```

```python
df = pd.DataFrame(Transactions_data, columns=items)
df['TransactionID'] = np.arange(1, NumOfTransactions + 1)
print(df)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_items_apriori = apriori(df.drop('TransactionID', axis=1), min_support=0.3, use_colnames=True)
frequent_items_apriori
```

```python
from mlxtend.frequent_patterns import fpgrowth
frequent_items_fptree = fpgrowth(df.drop('TransactionID', axis=1), min_support=0.3, use_colnames=True)
frequent_items_fptree
```

```python
rules_apriori = association_rules(frequent_items_apriori, metric="confidence", min_threshold=0.5)
rules_fptree = association_rules(frequent_items_fptree, metric="confidence", min_threshold=0.5)
```

```python
comparison_df = pd.DataFrame({
    'Algorithm': ['Apriori', 'FP-Growth'],
    'Frequent Itemsets': [len(frequent_items_apriori), len(frequent_items_fptree)],
    'Association Rules': [len(rules_apriori), len(rules_fptree)]
})
comparison_df
```
